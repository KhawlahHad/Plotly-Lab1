## Q1: Combine the two datsets as one dataframe

```python
movies_credits = movies.merge(credits, left_on='id', right_on='movie_id', how='inner')
```

## Q2: Remove the following columns:
- 'homepage','tagline','id','overview','status','original_title','movie_id'.

```python
movies_credits = movies_credits.drop(columns=['homepage','tagline','id','overview','status','original_title','movie_id'])
```

```python
movies.head(3)
```

## Q3: Create Genres dataframe that hase two coulumns Genres and Frequency
- Hint: Use Counter to calculate Frequency
- Add new catagory to Genres called Others when the Frequency < 200

```python

genres_list =[]

def add_names(str_lst):
    lst = json.loads(str_lst)
    for item in lst:
        genres_list.append(item['name'])


movies_credits['genres'].apply(add_names)

genre_counter = Counter(genres_list)

Genres = pd.DataFrame(list(genre_counter.items()), columns=['Genres', 'Frequency'])

Genres['Genres'] = Genres['Genres'].apply(lambda name: 'Other' if genre_counter[name] < 200 else name)

print(Genres)
```

## Q4: Use bar chart draws the relationship between genres of movies and Frequency.


```python
barchart = px.bar(Genres, x='Genres', y='Frequency', title='Genres of Movies',hover_data=['Frequency', 'Genres'], color='Genres')

barchart.show()
```

## Q5: Use pie chart to draw top 5 languages.
- Hint: use counter to create dataframe contains production_countries with Frequencies.


```python
prod_countries_list =[]

def add_countries(str_lst):
    lst = json.loads(str_lst)
    for item in lst:
        prod_countries_list.append(item['name'])


movies_credits['production_countries'].apply(add_countries)

prod_country_counter = Counter(prod_countries_list)

production_countries_df = pd.DataFrame(list(prod_country_counter.most_common(5)), columns=['Country', 'Frequency'])


print(production_countries_df)

fig = px.pie(production_countries_df, values='Frequency', names='Country', title='Production Countries')
fig.show()
```

## Q6: Use WordCloud to draw genres.

```python
word_freq_dict = dict(zip(Genres['Genres'], Genres['Frequency']))
wordcloud = WordCloud(width=400, height=200, background_color="white").generate_from_frequencies(word_freq_dict)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
